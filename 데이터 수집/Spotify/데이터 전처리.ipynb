### audio feature 추출

#### 2020년

```python
af_2020 = pd.read_excel(path + "/audio_features_2020.xlsx", index_col=0)
```

```python
af_2020
```

```python
af_2020.fillna('0', inplace=True)
```

```python
eval(af_2020.iloc[0, 0])
```

```python
af_2020.shape
```

```python
af_2020_new = pd.DataFrame()
af = ['id', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
      'instrumentalness', 'liveness', 'valence', 'tempo']

for i in range(1000):
    for j in range(41):
        if af_2020.iloc[i,j] == '0':
            continue
        af_2020_new = pd.concat((af_2020_new, pd.DataFrame(eval(af_2020.iloc[i,j]), index=[0]).loc[:, af]))
```

```python
af_2020_new.reset_index(drop=True, inplace=True)
```

```python
af_2020_new
```

```python
print(af_2020_new['id'].duplicated().sum())
print(af_2020_new['id'].isnull().sum())
```

```python
af_2020_final = af_2020_new.drop_duplicates(keep='first').reset_index(drop=True)
```

```python
af_2020_final
```
