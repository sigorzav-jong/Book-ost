# API로부터 각 노래의 Audio feature 열 추가

```python
song = pd.read_csv(path + '/audio feature 감정 추출 모델/to_make_af.csv',index_col=0 , encoding='utf-8')
song
```

```python
af_list = []
for i in song['id']:
    af_list.append(sp.audio_features(i))
  
af_list[:2]
```

```python
song['af'] = af_list
print(song.shape)
song.head()
```

```python
song['af'][0][0]
```

```python
danceability =[]
energy =[]
key = []
loudness = []
mode =[]
speechiness =[]
acousticness =[]
instrumentalness =[]
liveness =[]
valence =[]
tempo =[]
```

```python
for i in song['af']:
    danceability.append(i[0]['danceability'])
    energy.append(i[0]['energy'])
    key.append(i[0]['key'])
    loudness.append(i[0]['loudness'])
    mode.append(i[0]['mode'])
    speechiness.append(i[0]['speechiness'])
    acousticness.append(i[0]['acousticness'])
    instrumentalness.append(i[0]['instrumentalness'])
    liveness.append(i[0]['liveness'])
    valence.append(i[0]['valence'])
    tempo.append(i[0]['tempo'])
```

```python
af_df = pd.DataFrame([danceability,energy,key,loudness,mode,speechiness,acousticness,instrumentalness,liveness,valence,tempo],index = ['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo']).T
print(af_df.shape)
af_df.head(5)
```

```python
song = pd.concat([song, af_df],axis=1)
song.head()
```

```python
song.drop('af',axis=1,inplace=True)
print(song.shape)
song.head()
```

```python
song['책제목'].unique()
```

```python
def mood(x):
    if x == '두고 온 여름':
        return ["empty", "sadness", "worry", "boredom"]
    elif x == '백일의 밤 백편의 시':
        return ["sadness", "love", "happiness", "relief"]
    elif x== '어른을 위한 인생 수업  류쉬안':
        return ["happiness"]
    elif x == ' 쓸데없는 짓이 어디 있나요':
        return ["fun", "happiness", "relief"]
    elif x== '스키터':
        return ['relief']
    elif x== '잊었던 용기':
        return ["relief", "happiness", "love"]
    elif x== '다른 삶':
        return ["worry", "empty"]
    elif x =='당연한 하루는 없다':
        return ["worry", "relief"]
    elif x == '엉망인 채 완전한 축제':
        return ["sadness", "empty", "worry"]
    elif x =='작은 별이지만 빛나고 있어':
        return ["happiness", "love"]
    elif x =='달러구트 꿈 백화점':
        return ["enthusiasm", "happiness"]
    elif x == '어린 왕자':
        return ["happiness", "love", "fun"]
    elif x =='데미안':
        return ["worry", "sadness"]
    elif x =='이방인':
        return ["empty", "sadness"]
    elif x =='인간실격':
        return ["sadness", "empty"]
    elif x == '분노의 질주':
        return ["anger"]
    elif x == '헝거 게임':
        return ['anger','worry']
    elif x == '반지의 제왕':
        return ["enthusiasm", "worry"]
    elif x == '토이스토리':
        return ["fun", "love", "happiness"]
    elif x =='쿵푸 팬더':
        return ["fun", "enthusiasm"]
    elif x=='테이큰':
        return ["anger", "sadness","worry"]
    elif x =='아메리칸 뷰티':
        return ["boredom", "empty", "sadness"]
    elif  x =='로스트 인 트랜스레이션':
        return ["boredom", "empty", "sadness"]
    elif x == '로스트 인 트랜스 레이션':
        return ["boredom", "empty", "sadness"]
```

```python
song['mood'] = song['책제목'].apply(mood)
print(song.shape)
song.head()
```

```python
song.to_csv(path + '/audio feature 감정 추출 모델/traindata_af.csv')
```
