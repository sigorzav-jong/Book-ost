```python
import pandas as pd
import numpy as np
```

```python
keyword = pd.read_excel('song_keyword.xlsx',index_col=0) # 가사 전처리/ 가수 / 노래제목/ url 포함
lyrics = pd.read_excel('lyrics_sentiment.xlsx',index_col=0) # 가사기반 감정 확률
song = pd.read_excel('book_ost_song_final_data.xlsx',index_col=0) # audio_feature 기반 감정 확률
```

```python
print(song.shape)
song.head()
```

```python
song.columns = ['id','name','Artist','empty_a', 'sadness_a', 'enthusiasm_a', 'worry_a', 'love_a', 'fun_a',
       'hate_a', 'happiness_a', 'boredom_a', 'relief_a', 'anger_a']
song.head()
```

```python
song_drop_index = song[song[['name','Artist']].duplicated()].index
song.drop(song_drop_index,inplace=True,axis=0)
song
```

```python
song = song.reset_index().iloc[:,1:]
song
```
